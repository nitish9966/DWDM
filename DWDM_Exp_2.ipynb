```python
import pandas as pd
import numpy as np

cancer=pd.read_csv("data.csv")
print(cancer.head())
```

```python
data=pd.DataFrame(cancer)
```

```python
data['radius_category']=pd.cut(data['radius_mean'],bins=[0,10,20,30,np.inf],labels=['0-10','10-20','20-30','>30'])
print(data['radius_category'])
print(data.head())
```

```python
X_data=data.drop('radius_category',axis=1).values
y=data['radius_category']
print(X_data)
print(y)
```
